# NWS And Twitter API's 
The documentation for the National Weather Service (NWS) API are in this link 
- https://www.weather.gov/documentation/services-web-api
The documentation for the Twitter API are in this link 
- https://developer.twitter.com/en/docs

```python
#Import Libraries 
import pandas as pd 
import requests
import sqlite3
nws = 'https://api.weather.gov/alerts/active'
conn = sqlite3.connect('example.db')
```

```python
payload = {'status':'actual',
           'message_type': 'alert',
           'severity':'severe,extreme'
}

res = requests.get(nws, params=payload)
```
